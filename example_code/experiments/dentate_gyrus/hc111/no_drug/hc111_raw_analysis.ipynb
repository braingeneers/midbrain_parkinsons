# Set up

```python
import pickle
from matplotlib import pyplot as plt
from human_hip.raw_data import get_brain_waves, cwt_spectrogram, spectrogram_plot
from human_hip.spike_data import read_autocurated_data, plot_raster, read_phy_files
```

```python
sd = read_phy_files( "/workspaces/human_hippocampus/data/ephys/2022-11-02-e-Hc11.1-chip16753/derived/kilosort2/hc11.1_plated-11.1_chip-16753_rec_11.2.22AM-immediate_curated.zip" )
```

```python
plot_raster( sd )
```

# spectrogram - 34 seconds

```python
plot_raster(sd, xlim=(34,44), ylim=(0, .5) )
```

```python
with open("/workspaces/human_hippocampus/data/ephys_raw/2022-11-02-e-Hc11.1-chip16753/immediate_start34s_stop44s_fs1ms.pkl", 'rb') as filename:
    data_raw = pickle.load(filename)
print(data_raw.keys())
print(data_raw["data"].shape)
```

```python
plt.scatter( data_raw["xy"][:,0] , data_raw["xy"][:,1], c="blue" )
plt.scatter( data_raw["xy"][18,0] , data_raw["xy"][18,1], c="red" )
```

```python
data = get_brain_waves( data_raw["data"], data_raw["frame_rate"] )#[wave_type]  # get common brain waves
```

```python
power, times, frequencies, coif = cwt_spectrogram( data["basic"][18,:],  data_raw["frame_rate"], nNotes=12)
```

```python
#print(i, channel,raw_trace.shape)
fig, axs = plt.subplots(1, 1, figsize=(20, 6))
spectrogram_plot(power, times, frequencies, coif, ax=axs)

axs.set_title(f"spectrum", fontsize=12)
#axs.set_xlim(0, len(data1_lfp)/fs ) # code causes errors
#axs.set_ylim(0, 100)
axs.set_xlabel('time (s)')
axs.set_ylabel('frequency (Hz)')
axs.set_yscale('log')
plt.show(block=False)
```
